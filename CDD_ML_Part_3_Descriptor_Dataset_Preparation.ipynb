## **Calculate fingerprint descriptors**


### **Calculate PaDEL descriptors**

```
! wget https://github.com/dataprofessor/padel/raw/main/fingerprints_xml.zip
! unzip fingerprints_xml.zip
```

```
#List and sort fingerprint XML files
import glob
xml_files = glob.glob("*.xml")
xml_files.sort()
xml_files
```

```

FP_list = ['AtomPairs2DCount',
 'AtomPairs2D',
 'EState',
 'CDKextended',
 'CDK',
 'CDKgraphonly',
 'KlekotaRothCount',
 'KlekotaRoth',
 'MACCS',
 'PubChem',
 'SubstructureCount',
 'Substructure']
```

```
fp = dict(zip(FP_list, xml_files))
fp
```

```
from padelpy import padeldescriptor

# Create input file for PaDEL
input_file = "molecules.smi"

df3[['canonical_smiles','molecule_chembl_id']].to_csv(
    input_file,
    sep='\t',
    header=False,
    index=False
)
```

```
output_file = "descriptors_output.csv"

padeldescriptor(
    mol_dir=input_file,
    d_file=output_file,
    fingerprints=True,
    removesalt=True
)
```
